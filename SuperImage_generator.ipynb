```python
%matplotlib inline
import sys
from PIL import Image

# image_1 = Image.open(os.path.join(os.getcwd(),'images','bird.jpg'))
# image_1.resize((224,224))
image_2 = Image.open(os.path.join(os.getcwd(),'images','dog.jpg'))
print(image_2.size)
image_2 = image_2.resize((224,224))
print(image_2.size)
plt.imshow(np.squeeze(image_2))
plt.show()
```

```python
def horizontal_align(images):
    Images = [Image.open(os.path.join(os.getcwd(),'images',x)) for x in images]
    new_images = []
    for img in Images:
        img = img.resize((112,112))
        new_images.append(img)
    new_im = Image.new('RGB', (224,112))
    x_offset = 0
    for im in new_images:
        new_im.paste(im,(x_offset,0))
        x_offset += 112
    return new_im

def vertical_align(images):
    new_im = Image.new('RGB', (224,224))
    
    y_offset = 0
    for img in images:
        new_im.paste(img,(0,y_offset))
        y_offset += img.size[1]
    
    return new_im
```

```python
line_1 = horizontal_align( ['cat.jpg', 'bird.jpg'])
line_2 = horizontal_align( ['fruits.jpg', 'water-bird.JPEG'])

final_image = vertical_align([line_1,line_2])
final_image.save(os.path.join(os.getcwd(),'images',"super4.jpg"))
plt.imshow(final_image)
```
